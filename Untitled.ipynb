```python
from math import sqrt
from scipy.stats import norm
import numpy as np
```

```python
def brownian(x0, ts, delta, out=None):
    """
    Generate an instance of Brownian motion (i.e. the Wiener process):

        X(t) = X(0) + N(0, delta**2 * t; 0, t)

    where N(a,b; t0, t1) is a normally distributed random variable with mean a and
    variance b.  The parameters t0 and t1 make explicit the statistical
    independence of N on different time intervals; that is, if [t0, t1) and
    [t2, t3) are disjoint intervals, then N(a, b; t0, t1) and N(a, b; t2, t3)
    are independent.
    
    Written as an iteration scheme,

        X(t + dt) = X(t) + N(0, delta**2 * dt; t, t+dt)


    If `x0` is an array (or array-like), each value in `x0` is treated as
    an initial condition, and the value returned is a numpy array with one
    more dimension than `x0`.

    Arguments
    ---------
    x0 : float or numpy array (or something that can be converted to a numpy array
         using numpy.asarray(x0)).
        The initial condition(s) (i.e. position(s)) of the Brownian motion.
    n : int
        The number of steps to take.
    dt : float
        The time step.
    delta : float
        delta determines the "speed" of the Brownian motion.  The random variable
        of the position at time t, X(t), has a normal distribution whose mean is
        the position at time t=0 and whose variance is delta**2*t.
    out : numpy array or None
        If `out` is not None, it specifies the array in which to put the
        result.  If `out` is None, a new numpy array is created and returned.

    Returns
    -------
    A numpy array of floats with shape `x0.shape + (n,)`.
    
    Note that the initial value `x0` is not included in the returned array.
    """

    x0 = np.asarray(x0)
    ts = np.asarray(ts)
    delta_ts = np.zeros_like(ts)
    delta_ts[:, 0] = ts[:, 0]
    delta_ts[:, 1:] = ts[:, 1:] - ts[:, :-1] 
    
    # For each element of x0, generate a sample of n numbers from a
    # normal distribution.
    r = norm.rvs(size=x0.shape + (delta_ts.shape[-1],), scale=delta*np.sqrt(delta_ts))

    # If `out` was not given, create an output array.
    if out is None:
        out = np.empty(r.shape)

    # This computes the Brownian motion by forming the cumulative sum of
    # the random samples. 
    np.cumsum(r, axis=-1, out=out)

    # Add the initial condition.
    out += np.expand_dims(x0, axis=-1)
    print(x0)

    return out
```

```python
def ou(x0, ts, theta, sigma, mu, out=None):
    x0 = np.asarray(x0)
    ts = np.asarray(ts)
    delta_ts = np.zeros_like(ts)
    delta_ts[:, 0] = 1 - np.exp(-2*theta*ts[:, 0])
    delta_ts[:, 1:] = 1- np.exp(-2*theta*(ts[:, 1:] - ts[:, :-1]))
    # delta_ts[0] = 1 - np.exp(-2*theta*ts[0])
    # delta_ts[1:] = 1 - np.exp(-2*theta*(ts[1:] - ts[:-1] ))
        
    # For each element of x0, generate a sample of n numbers from a
    # normal distribution.
    r = norm.rvs(size=x0.shape + (delta_ts.shape[-1],), scale=np.sqrt(delta_ts))

    # If `out` was not given, create an output array.
    if out is None:
        out = np.empty(r.shape)

    # This computes the Brownian motion by forming the cumulative sum of
    # the random samples. 
    # cumsum = np.cumsum(r, axis=-1)
    out[:, 0] = x0* np.exp(-theta*ts[:,0]) + mu * (1-np.exp(-theta*ts[:,0])) + sigma/np.sqrt(2*theta) * r[:,0]
    # print(out[:, 0])
    for i in np.arange(out.shape[-1]-1):
        timediff = ts[:,i+1] - ts[:, i]
        out[:, i+1] = out[:, i] * np.exp(-theta*timediff) + mu * (1-np.exp(-theta*timediff)) + sigma/np.sqrt(2*theta) * r[:,i]

    # print(out)

    return out

def oua(x0, ts, theta, sigma, mu, out=None):
    x0 = np.asarray(x0)
    ts = np.asarray(ts)
    delta_ts = np.zeros_like(ts)
    delta_ts[:, 0] = 1 - np.exp(-2*theta*ts[:, 0])
    delta_ts[:, 1:] = 1- np.exp(-2*theta*(ts[:, 1:] - ts[:, :-1]))
    # delta_ts[0] = 1 - np.exp(-2*theta*ts[0])
    # delta_ts[1:] = 1 - np.exp(-2*theta*(ts[1:] - ts[:-1] ))
        
    # For each element of x0, generate a sample of n numbers from a
    # normal distribution.
    r = norm.rvs(size=x0.shape + (delta_ts.shape[-1],), scale=sigma/np.sqrt(2*theta) * np.sqrt(delta_ts))

    # If `out` was not given, create an output array.
    if out is None:
        out = np.empty(r.shape)

    # This computes the Brownian motion by forming the cumulative sum of
    # the random samples. 
    np.cumsum(r, axis=-1, out=out)
    # out = sigma/np.sqrt(2*theta) * out

    # Add the initial condition.
    # print( np.expand_dims(x0, axis=-1) * np.exp(-theta*ts), )
    out += np.expand_dims(x0, axis=-1) * np.exp(-theta*ts) + mu * (1-np.exp(-theta*ts))

    return out

def outrans(x0, ts, theta, sigma, mu, out=None):
    x0 = np.asarray(x0)
    ts = np.asarray(ts)
    delta_ts = np.zeros_like(ts)
    delta_ts[:, 0] = ts[:, 0]
    delta_ts[:, 1:] = ts[:, 1:] - ts[:, :-1]
    frac = (1 - np.exp(-2*theta*ts[:, 1:]) ) / ts[:, 1:]
    frac[frac < 0] = 0.
    scale = np.zeros_like(ts)
    scale[:, 1:] = sigma/np.sqrt(2*theta) * np.sqrt(frac)
    # delta_ts[0] = 1 - np.exp(-2*theta*ts[0])
    # delta_ts[1:] = 1 - np.exp(-2*theta*(ts[1:] - ts[:-1] ))
        
    # For each element of x0, generate a sample of n numbers from a
    # normal distribution.
    r = norm.rvs(size=x0.shape + (delta_ts.shape[-1],), scale=np.sqrt(delta_ts))
    # print(r)

    # If `out` was not given, create an output array.
    if out is None:
        out = np.empty(r.shape)

    # This computes the Brownian motion by forming the cumulative sum of
    # the random samples. 
    np.cumsum(r, axis=-1, out=out)
    out = scale * out

    # Add the initial condition.
    # print( np.expand_dims(x0, axis=-1) * np.exp(-theta*ts), )
    out += np.expand_dims(x0, axis=-1) * np.exp(-theta*ts) + mu * (1-np.exp(-theta*ts))

    return out

def ou2(x0, ts, theta, sigma, mu, out=None):
    x0 = np.asarray(x0)
    ts = np.asarray(ts)
    delta_ts = np.zeros_like(ts)
    delta_ts[:, 0] = ts[:, 0]
    delta_ts[:, 1:] = ts[:, 1:] - ts[:, :-1]
    # delta_ts[0] = 1 - np.exp(-2*theta*ts[0])
    # delta_ts[1:] = 1 - np.exp(-2*theta*(ts[1:] - ts[:-1] ))
    
    # print(delta_ts)
    
    # For each element of x0, generate a sample of n numbers from a
    # normal distribution.
    r = np.exp(theta * ts) * norm.rvs(size=x0.shape + (delta_ts.shape[-1],), scale=np.sqrt(delta_ts))

    # If `out` was not given, create an output array.
    if out is None:
        out = np.empty(r.shape)

    # This computes the Brownian motion by forming the cumulative sum of
    # the random samples. 
    np.cumsum(r, axis=-1, out=out)
    out *= sigma * np.exp(-theta * ts)

    # Add the initial condition.
    # print( np.expand_dims(x0, axis=-1) * np.exp(-theta*ts), )
    out += np.expand_dims(x0, axis=-1) * np.exp(-theta*ts) + mu * (1-np.exp(-theta*ts))

    return out

def ou3(x0, ts, theta, sigma, mu, out=None):
    x0 = np.asarray(x0)
    ts = np.asarray(ts)
    delta_ts = np.zeros_like(ts)
    delta_ts[:, 0] = np.exp(2*theta*ts[:, 0]) - 1
    delta_ts[:, 1:] = np.exp(2*theta*(ts[:, 1:] - ts[:, :-1])) - 1
    # delta_ts[0] = 1 - np.exp(-2*theta*ts[0])
    # delta_ts[1:] = 1 - np.exp(-2*theta*(ts[1:] - ts[:-1] ))
        
    # For each element of x0, generate a sample of n numbers from a
    # normal distribution.
    r = norm.rvs(size=x0.shape + (delta_ts.shape[-1],), scale=sigma/np.sqrt(2*theta) * np.sqrt(delta_ts))
    
    print(r)

    # If `out` was not given, create an output array.
    if out is None:
        out = np.empty(r.shape)

    # This computes the Brownian motion by forming the cumulative sum of
    # the random samples. 
    np.cumsum(r, axis=-1, out=out)
    print(np.exp(-theta*ts))
    print(delta_ts)
    print(out)
    out = np.exp(-theta*ts) * out
    print(out)

    # Add the initial condition.
    # print( np.expand_dims(x0, axis=-1) * np.exp(-theta*ts), )
    out += np.expand_dims(x0, axis=-1) * np.exp(-theta*ts) + mu * (1-np.exp(-theta*ts))
    
    return out
```
