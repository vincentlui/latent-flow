```python
class OrnsteinUhlenbeckSDE(torch.nn.Module):
    sde_type = 'ito'
    noise_type = 'scalar'

    def __init__(self, mu, theta, sigma):
        super().__init__()
        self.register_buffer('mu', torch.as_tensor(mu))
        self.register_buffer('theta', torch.as_tensor(theta))
        self.register_buffer('sigma', torch.as_tensor(sigma))

#     def f(self, t, y):
#         return self.mu * t - self.theta * y

#     def g(self, t, y):
#         return self.sigma.expand(y.size(0), 1, 1) * (2 * t / 10)
        
    def f(self, t, x):
        sigma = 10.
        ro = 28.
        beta = 8./3.
        out = torch.empty_like(x)
        out[0] = sigma * (x[1] - x[0])
        out[1] = x[0] * (ro - x[2]) - x[1]
        out[2] = x[0] * x[1] - beta * x[2]
        return out

    def g(self, t, x):
        ax = 0.1
        ay = 0.28
        az = 0.3
        return torch.Tensor([[[ax, ay, az]]])

ou_sde = OrnsteinUhlenbeckSDE(mu=0.02, theta=0.1, sigma=0.4)
y0 = torch.rand(1).unsqueeze(-1) * 2 - 1
ts = torch.linspace(0, 10 - 1, 10)
ys = torchsde.sdeint(ou_sde, y0, ts, dt=1e-1)
```
